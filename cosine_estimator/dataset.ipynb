```python
import numpy as np
import pandas as pd
import json
```

```python
data = {'y':[], 'x':[]}
for x in range(100):
    x1  = x/50
    x2  =  x/20
    data['x'].append( [x1, x2]  )

    y = np.cos(x1) + x2
    data['y'].append(y)

open('data.json','w').write( json.dumps( [data['y'], data['x']], indent=2) )

data['y']
```
